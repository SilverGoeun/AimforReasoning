```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```
x= np.linspace (-5, 5, 101)
y= np.sin (x)

plt.plot (x , y)
```

그래프를 인라인에 출력

```
fix, ax = plt.subplots()

ax.set_title ('Sin graph')
ax.set_xlabel('radian')
ax.plot (x, y)

handles, labels = ax.get_legend_handles_labels()
ax.legend (handles, labels)
plt.show()
```
