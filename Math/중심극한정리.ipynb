```
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(0)
%precision 3
%matplotlib inline
```

포아송 분포

```
l = 3
rv = stats.poisson (l)

n = 1000
sample_size = 10000
### 혹시 random state와 연관이 있나????
###Xs_sample = rv.rvs ( ( n, sample_size, random_state = ?? None아닌 걸로))
Xs_sample = rv.rvs ( ( n, sample_size))
sample_mean = np.mean ( Xs_sample, axis = 0)

rv_true = stats.norm ( l, np.sqrt (1/n))
xs = np.linspace (rv_true.isf(0.999), rv_true.isf(0.001),100)
```

```
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)

ax.hist( sample_mean, bins=100, density = True,
          alpha = 0.5, label = ' sample mean of 10000* Poi(3) ')
ax.plot ( xs, rv_true.pdf(xs), label = 'N( 3, 3/10000)', color ='gray')

ax.legend()
ax.set_xlim(rv_true.isf(0.999), rv_true.isf(0.001))
plt.show()
```

원문복사

```
l = 3
rv = stats.poisson(l)

n = 1000
sample_size = 10000
Xs_sample = rv.rvs((n, sample_size))
sample_mean = np.mean(Xs_sample, axis=0)

rv_true = stats.norm(l, np.sqrt(l/n))
xs = np.linspace(rv_true.isf(0.999), rv_true.isf(0.001), 100)
```

```

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)

ax.hist(sample_mean, bins=100, density=True,
        alpha=0.5, label='10000個のPoi(3)の標本平均')
ax.plot(xs, rv_true.pdf(xs), label='N(3, 3/10000)', color='gray')

ax.legend()
ax.set_xlim(rv_true.isf(0.999), rv_true.isf(0.001))
plt.show()
```

균등분포 (Uniform Distribution)

```
l = 10
rv = stats.uniform (l)

n = 5
sample_size = 1000
Xs_sample = rv.rvs ( (n,  sample_size ))
sample_mean = np.mean ( Xs_sample, axis =0 )

rv_true = stats.norm (l, np.sqrt (1/n))
xs = np.linspace (rv_true.isf(0.999), rv_true.isf(0.001), 100)
```

```
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)

ax.hist(sample_mean, bins=100, density=True,alpha = 0.5,
         label='sample mean of n number of Uniform(l)')
ax.plot(xs, rv_true.pdf(xs), label='N(l, l/n)', color='gray')

ax.legend()
ax.set_xlim(rv_true.isf(0.999), rv_true.isf(0.001))
plt.show()
```
