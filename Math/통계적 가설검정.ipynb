#** 정규분포의 모분산에 대한 검정**

```
def pvar_test ( sample, var0 , alpha = 0.05):
    u_var = np.var (sample, ddof=1)
    n = len (sample)
    rv = stats.chi2 (df = n-1 )
    interval = rv.interval (1 - alpha)

    y = ( n-1 ) * u_var / var0
    if interval [0] <= y <= interval [1]:
        print ('귀무가설을 채택')
    else :
        print  ('구무가설을 기각')
    if y < rv.isf(0.5) :
        p = rv.cdf(y) *2
    else:
        p = ( 1 - rv.cdf(y)) * 2
    print (f'p값은 {p:.3f}')
    
```

```
# 분산이 9이지 검정
pvar_test (sample, 9)
```
