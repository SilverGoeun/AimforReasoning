# **유방암 진단 데이터세트**

```
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()

X = data.data
y = data.target

X = X[ :, :7]
```

```
X
```

유방암 로지스틱 회귀

```
from sklearn.linear_model import LogisticRegression

model = LogisticRegression (solver = 'lbfgs')

model.fit ( X, y)
```

```
y_pred = model.predict (X)
```

```
from sklearn.metrics import accuracy_score

accuracy_score ( y, y_pred)
```

# **와인 k-clustering**

```
from sklearn.datasets import load_wine

data = load_wine()
```

```
X= data.data[ :, [0,9]]
len (X)
```

```
from sklearn.cluster import KMeans

n_cluster = 3
model = KMeans (n_clusters =n_cluster)

pred = model.fit_predict (X)
```

```
%matplotlib inline
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.scatter ( X[pred ==0, 0], X[pred ==0, 1], color = 'red', marker= 's', label= 'Label1')
ax.scatter(X[pred==1, 0], X[pred==1, 1], color='blue', marker='s', label='Label2')
ax.scatter(X[pred==2, 0], X[pred==2, 1], color='green', marker='s', label='Label3')
ax.scatter ( model.cluster_centers_[:, 0], model.cluster_centers_[:,1], s=200, color ='yellow',
             marker='*', label="center"  )

ax.legend()
plt.show()
```
