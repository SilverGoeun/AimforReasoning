```
c = kolaw.open ('constitution.txt').read()
```

```
con = c.split ('\n\n')
con[0]
```

```
con_sen = con[1]
print(con_sen)
```

```
#꼬꼬마
#형태소를 분리하는 모듈로 형태소 객체 생성
from konlpy.tag import Kkma
kkma = Kkma()
```

```
kkma.sentences(con_sen)
#문장을 분리함. 해당 헌법 문장은 한 문장이라 리스트 내에 하나의 문자열만 표시됨
```

```
#명사 기준 분리 후 10개만 확인
con_sen_noun = kkma.nouns (con_sen)
con_sen_noun[:10]
```

```
#한글을 형태소별로 분리
con_sen_m = kkma.morphs (con_sen)
con_sen_m [:10]
```

```
#품사 분리
con_sen_p = kkma.pos (con_sen)
con_sen_p [:10]
```

```
#한나눔 형태소 분석기
from konlpy.tag import  Hannanum
hannanum = Hannanum()
```

```
hannanum.nouns ('이것은 형태소 분석기 예제입니다. 아버지가 방에 들어가신다고 해 봅시다.')
```

```
hannanum.morphs ('형태소 분석기 예제입니다. 아버지가 방에 들어가신다고 해 봅시다.')
```

```
hannanum.pos ('형태소 분석기 예제입니다. 아버지가 방에 들어가신다고 해 봅시다.')
```

```
# Twitter 형태소 분석기 Open Korean Text (Okt)
from konlpy.tag import Okt
okt = Okt()
okt.nouns ('이것은 형태소 분석기 예제입니다. 아버지가 방에 들어가신다고 해 봅시다.')
```

```
okt.pos ('이것은 형태소 분석기 예제입니다. 아버지가 방에 들어가신다고 해 봅시다.')
```

```
text = '이것은 형태소 분석기 예제입니다. 아버지가 방에 들어가신다고 해 봅시다.'
print (okt.morphs (text))
#어간을 만듦
print (okt.morphs (text, stem = True)) 
```

```
#관용구 처리, 특정 구별로 분리
okt.phrases (text)
```
