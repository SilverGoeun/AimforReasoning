```
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```
x = tf.constant(1.)
bool = tf.constant(True)
res = tf.cond(bool, lambda: tf.add(x, 1.), lambda: tf.add(x, 10.))

with tf.Session():
    result = res.eval()
print(result)
```
